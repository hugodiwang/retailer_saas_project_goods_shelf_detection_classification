In this part, we will package our trained model into a SAAS web service. We select a python based plateform flask. And ORM will be used to connect class object with database.

step 1:
download our dependancy requirements:

cd drive/MyDrive/retailer_detect/ai_saas/ai_saas

pip install -r requirements.txt 

step 2: start flask

 python manage.py (dev version)

step 3: start a post request

cd body json file
and then post

curl -H "Content-Type: application/json" --data @body.json http://localhost:9000/ai_saas


step 4: write docker file
based on tensorflow/tensorflow:2.2.0-gpu image
here we used a gpu version image. If using cpu version, drop these words.

step 5: build docker image

docker build -t tf2-ai-saas -f ai_saas/Dockerfile .


step 6: run gpu version docker image
docker run --runtime nvidia -it --rm --name tf2_ai_saas -p 9000:9000 tf2-ai-saas bash

--rm if crash, image will be removed
--runtime nvidia, gpu image
